```python
def count(df,spark):
    my_count = df.count()
    return my_count
```

```python
def getNumberOfFields(df,spark):
    num_fields = len(df.columns)
    return num_fields
```

```python
def getFieldNames(df,spark):
    field_names = list(df.columns)
    return field_names
```
